```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('homicide_by_countries.csv')
```

```python
df['Rate'] = df['Rate'].astype(int)
```

```python
my_list = ['Rate','Count','Year']
for i in my_list:
  print(i)
```

```python
df
```

```python
df1 = df.sort_values(by='Count',ascending=False).reset_index(drop = True)
df1
```

```python
df1 = df[['Location','Count']].sort_values(by='Count',ascending=False).head()
df1['Perc'] = (df1['Count'] * 100 / df1['Count'].sum()).round(2)
df1
```

```python
df1.Location
```

```python
df1.plot(x = 'Location', y = 'Count', kind = 'pie', labels = df1.Location, autopct='%.2f')
plt.legend().set_visible(False)
plt.show()
```

```python
df['Region'].replace('Americas','N/S America', inplace=True)
```

```python
df2 = df.groupby('Region')['Count'].sum().sort_values(ascending=False)
df2
```

```python
df2.plot(kind = 'bar', color = 'g')
plt.show()
```

```python
df3 = df.groupby('Subregion')['Count'].sum().sort_values(ascending = False)
df3
```

```python
sns.barplot(x = df3.index, y = df3.values , hue = df3.index)
plt.xticks(rotation = 90)
plt.show()
```

```python
df.Year.value_counts()
```

```python
df[df['Region'].isin(['Asia','Europe'])]
```

```python
df4 = df[(df['Region'] == 'Asia') | (df['Region'] =='Europe')]
df4
```

```python
df4 = df4[df4['Year'] > 2016][['Region','Year','Count']]
```

```python
df4 = df4.groupby(['Region','Year']).sum()['Count']
df4
```

```python
df_unstacked = df4.unstack(level = 0)
df_unstacked
```

```python
df_unstacked.index = df_unstacked.index.astype(int).astype(str)
df_unstacked
```

```python
df_unstacked.plot(kind = 'line', figsize = (10,6))
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Homicide Count in Asia and Europe')
plt.show()
```

```python
df5 = df.groupby(['Year'])['Rate'].sum().sort_values(ascending = False)
df5
```

```python
df5.plot(kind = 'bar',figsize = (7,3), color = 'skyblue', edgecolor = 'black')
plt.xlabel('Year')
plt.ylabel('Rate')
plt.title('Homicide Rate by Year')
plt.show()
```

```python
df6 = df[['Year','Region','Count']]
df6 = df6.groupby(['Year','Region']).sum().sort_values(by = 'Year', ascending = False).head(40)
df6
```

```python
df6.plot(kind = 'bar', figsize = (12,6), colormap = 'viridis')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Homicide Count by Year and Region')
plt.show()
```

```python
df7 = df.groupby('Subregion')['Count'].mean().sort_values(ascending = False).round(2)
df7
```

```python
df7.index
```

```python
df7.values
```

```python
data = {
    'Category' : df7.index,
    'Value' : df7.values,
    'Info': df7.values
}
df = pd.DataFrame(data)
df
```

```python
fig = px.treemap(df, path = ['Category'], values = 'Value',title = 'Treemap')
fig.update_traces(hovertemplate = 'Category: %{label}<br>value:{value}')
fig.show()
```
